```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

adr_outliers, adr_low, adr_high = detect_outliers_iqr(df, 'adr')
lead_outliers, lead_low, lead_high = detect_outliers_iqr(df, 'lead_time')

print(f"ADR outliers count: {len(adr_outliers)}")
print(f"Lead Time outliers count: {len(lead_outliers)}")
```

```python
df = df[(df['adr'] >= adr_low) & (df['adr'] <= adr_high)]
df = df[(df['lead_time'] >= lead_low) & (df['lead_time'] <= lead_high)]
```
